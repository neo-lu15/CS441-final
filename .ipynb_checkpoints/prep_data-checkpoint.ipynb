```python
import pandas as pd

df = pd.read_csv("nba_games_with_travel.csv")
df = df.sort_values(["PLAYER_ID","GAME_DATE"])

for stat in ["PTS","REB","AST"]:
    df[f"{stat}_LAST5"] = df.groupby("PLAYER_ID")[stat].rolling(5).mean().reset_index(0,drop=True)
    df[f"{stat}_LAST10"] = df.groupby("PLAYER_ID")[stat].rolling(10).mean().reset_index(0,drop=True)

df["HOME_GAME"] = df["MATCHUP"].str.contains("vs").astype(int)
df["BACK_TO_BACK"] = (df["DAYS_REST"]==1).astype(int)
df.fillna(0,inplace=True)

df.to_csv("ML_ready_dataset.csv",index=False)
print("dataset processed")
```
